```python
import nest_asyncio
import asyncio
from playwright.async_api import async_playwright

# Apply the nest_asyncio patch
nest_asyncio.apply()

# Define vc_urls as a global variable
vc_urls = []

async def scrape_vc_urls():
    base_url = "https://vc-mapping.gilion.com/venture-capital-firms/san-francisco?5825b8f1_page="
    global vc_urls  # Declare vc_urls as global to modify it within the function

    async with async_playwright() as p:
        browser = await p.chromium.launch()
        page = await browser.new_page()

        for page_num in range(1, 16):
            url = f"{base_url}{page_num}"
            await page.goto(url)
            print(f"Scraping {url}")

            # Collect all the URLs from the current page
            links = await page.query_selector_all('a.card-button.w-button')
            for link in links:
                href = await link.get_attribute('href')
                if href and href.startswith('/vc-firms/'):
                    full_url = f"https://vc-mapping.gilion.com{href}"
                    vc_urls.append(full_url)

        await browser.close()

    return vc_urls

async def main():
    await scrape_vc_urls()
    for url in vc_urls:
        print(url)

# Run the main function
await main()
```

```python
"""
next i want to iterate over all the urls i collected and collect the following data:

name and logo: <div class="vc-heading-container"><div id="w-node-cd88bbb9-77e6-e3b9-5543-09d7adf30f34-fce52eee" class="logo-top-container is--vc-card"><div class="vc-logo-container"><div class="html-embed-4 w-embed"><img src="https://logo.clearbit.com/http://www.anthemis.com" alt="Anthemis Group venture capital firm logo"></div></div><div class="vc-logo-container is--none w-condition-invisible"></div></div><h1 id="w-node-_1743f616-1944-660a-f988-fd98c0b7d554-fce52eee" class="left vc">Anthemis Group</h1></div> 
founded: <div class="flexbox-vertical"><div class="vc-info_category-title">Founded: </div><div class="numbers-on-cards">2010</div></div>
investments: <div class="flexbox-vertical"><div class="vc-info_category-title">Investments: </div><div class="numbers-on-cards">222</div></div>
exists: <div class="flexbox-vertical"><div class="vc-info_category-title">Exits: </div><div class="numbers-on-cards">19</div></div>
about: <div class="vc-info_container is--first extra-text"><div class="lesft w-embed"><h2 style="font-size:2em"> About Anthemis Group</h2></div><p class="sans padding_top">Anthemis Group is a global platform that cultivates change in the financial system by investing in, growing, and sustaining businesses.</p><div class="w-dyn-bind-empty w-richtext"></div></div>
portfolio companies: <div class="agency-info"><div class="vc-info_category-title">Portfolio companies: </div><div class="numbers-on-cards-left">Stable,Branch,Ivella,Happy Money,LocoNav,Premise Data,Maxwell Financial Labs,Cowbell Cyber,Axle,Daylight,Atomic,EasyHealth,Sydecar,OnSiteIQ,Alaffia Health,Trōv,Automatic,Matic Insurance,Kettle,Moven</div></div>
fund type: <div class="vc-info_tag-wrapper is--top"><div class="vc-info_category-title">Fund type:</div><div class="w-dyn-list"><div role="list" class="vc-collection-list-wrap w-dyn-items"><div role="listitem" class="w-dyn-item"><div class="found-type-tag"><div class="tag-text">Venture Fund</div></div></div></div></div></div>
investment stage: <div class="vc-info_tag-wrapper"><div class="vc-info_category-title">Investment stage:</div><div class="w-dyn-list"><div role="list" class="vc-collection-list-wrap w-dyn-items"><div role="listitem" class="w-dyn-item"><a href="https://vc-mapping.gilion.com/venture-capital-firms/early-stage-investors" class="w-inline-block"><div class="found-type-tag"><div class="tag-text">Seed</div></div></a></div><div role="listitem" class="w-dyn-item"><a href="https://vc-mapping.gilion.com/venture-capital-firms/series-a-investors" class="w-inline-block"><div class="found-type-tag"><div class="tag-text">Series A</div></div></a></div><div role="listitem" class="w-dyn-item"><a href="https://vc-mapping.gilion.com/venture-capital-firms/series-b-investors" class="w-inline-block"><div class="found-type-tag"><div class="tag-text">Series B</div></div></a></div><div role="listitem" class="w-dyn-item"><a href="#" class="w-inline-block"><div class="found-type-tag"><div class="tag-text">Series C</div></div></a></div><div role="listitem" class="w-dyn-item"><a href="#" class="w-inline-block"><div class="found-type-tag"><div class="tag-text">Series D</div></div></a></div><div role="listitem" class="w-dyn-item"><a href="https://vc-mapping.gilion.com/venture-capital-firms/early-stage-investors" class="w-inline-block"><div class="found-type-tag"><div class="tag-text">Pre-Seed</div></div></a></div></div></div></div>
investment focus: <div class="vc-info_tag-wrapper"><div class="vc-info_category-title">Investment focus:</div><div class="w-dyn-list"><div role="list" class="vc-collection-list-wrap w-dyn-items"><div role="listitem" class="w-dyn-item"><div class="found-type-tag"><div class="tag-text">FinTech</div></div></div><div role="listitem" class="w-dyn-item"><div class="found-type-tag"><div class="tag-text">InsurTech</div></div></div><div role="listitem" class="w-dyn-item"><div class="found-type-tag"><div class="tag-text">PropTech</div></div></div><div role="listitem" class="w-dyn-item"><div class="found-type-tag"><div class="tag-text">SaaS</div></div></div><div role="listitem" class="w-dyn-item"><div class="found-type-tag"><div class="tag-text">Software</div></div></div><div role="listitem" class="w-dyn-item"><div class="found-type-tag"><div class="tag-text">Developer Tools</div></div></div><div role="listitem" class="w-dyn-item"><div class="found-type-tag"><div class="tag-text">Big Data &amp; Analytics</div></div></div><div role="listitem" class="w-dyn-item"><div class="found-type-tag"><div class="tag-text">Enterprise</div></div></div><div role="listitem" class="w-dyn-item"><div class="found-type-tag"><div class="tag-text">Cybersecurity</div></div></div><div role="listitem" class="w-dyn-item"><div class="found-type-tag"><div class="tag-text">Artificial Intelligence &amp; Machine Learning (AI/ML)</div></div></div><div role="listitem" class="w-dyn-item"><div class="found-type-tag"><div class="tag-text">Apps</div></div></div></div></div></div>

sometimes its possible that for a certain field the data is not available this is okay.
"""
import asyncio
from playwright.async_api import async_playwright

async def scrape_vc_details(url):
    async with async_playwright() as p:
        browser = await p.chromium.launch()
        page = await browser.new_page()
        await page.goto(url)
        print(f"Scraping details from {url}")

        # Initialize data dictionary with default empty values
        data = {
            'url': url,
            'name': '',
            'logo': '',
            'founded': '',
            'investments': '',
            'exits': '',
            'about': '',
            'portfolio_companies': '',
            'fund_type': '',
            'investment_stage': '',
            'investment_focus': ''
        }

        # Helper function to safely extract text content
        async def safe_text_content(selector):
            try:
                return await page.text_content(selector, timeout=5000)  # Reduced timeout for faster failure recovery
            except Exception:
                return ''  # Return empty string if there's an error

        # Helper function to safely extract attribute content
        async def safe_get_attribute(selector, attribute):
            try:
                return await page.get_attribute(selector, attribute, timeout=5000)
            except Exception:
                return ''

        # Extract data using safe functions
        data['name'] = await safe_text_content('h1.left.vc')
        data['logo'] = await safe_get_attribute('div.html-embed-4.w-embed img', 'src')
        data['founded'] = await safe_text_content('div.vc-info_category-title:has-text("Founded:") + div.numbers-on-cards')
        data['investments'] = await safe_text_content('div.vc-info_category-title:has-text("Investments:") + div.numbers-on-cards')
        data['exits'] = await safe_text_content('div.vc-info_category-title:has-text("Exits:") + div.numbers-on-cards')
        data['about'] = await safe_text_content('div.vc-info_container.is--first.extra-text p.sans.padding_top')
        data['portfolio_companies'] = await safe_text_content('div.vc-info_category-title:has-text("Portfolio companies:") + div.numbers-on-cards-left')
        data['fund_type'] = await safe_text_content('div.vc-info_category-title:has-text("Fund type:") + div.w-dyn-list div.found-type-tag div.tag-text')
        data['investment_stage'] = await safe_text_content('div.vc-info_category-title:has-text("Investment stage:") + div.w-dyn-list div.found-type-tag div.tag-text')
        data['investment_focus'] = await safe_text_content('div.vc-info_category-title:has-text("Investment focus:") + div.w-dyn-list div.found-type-tag div.tag-text')

        await browser.close()
        return data

async def collect_vc_details():
    global vc_details  # Declare vc_details as global to modify it within the function
    for url in vc_urls:
        details = await scrape_vc_details(url)
        vc_details.append(details)
        print(details)

# Run the function to collect VC details
await collect_vc_details()

# Now vc_details is filled with data and can be used later in the notebook
```
